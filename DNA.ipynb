## What is DNA Origami?

DNA origami is like folding a long DNA string into tiny, exact shapes using short pieces that act like staples.  
This helps build super small structures with great precision.  
**For example:** scientists can make tiny boxes, smiley faces, or robots out of DNA!

## Why Crosslink?

Crosslinking means connecting the DNA strands more tightly, like gluing parts together.  
This helps by:  
- making the shapes stronger and more stable in the body (so they don’t fall apart easily),  
- protecting them from being broken down by enzymes,  
- making them work better for treatments like delivering medicine right where it’s needed.

**For example:** if a tiny DNA box carries medicine, crosslinking keeps it closed until it reaches the right spot in the body.

## How DNA Origami Works with pH

- At low pH (like in tumors), DNA origami shows special parts outside.  
- These parts stick to cell targets and help kill bad cells.  
- At normal pH, the parts stay hidden and don’t work.  
- More DNA origami enters cells when pH is low.

```python
df = pd.read_csv('C:/Kaggle/NewProject2025/RoboticDNA.csv', header=None, names=["Measure", "TumorVolume_cm3"])
```

```python
df
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(df['Measure'], df['TumorVolume_cm3'], s=10, alpha=0.6)
plt.title("Tumor Growth Data")
plt.xlabel("Measure")
plt.ylabel("Tumor Volume (cm³)")
plt.grid(True)
plt.show()
```

```python
series_len = len(df) // 3

pbs = df.iloc[0:series_len].sort_values(by=["Measure", "TumorVolume_cm3"])
peg_o6p = df.iloc[series_len:2*series_len].sort_values(by=["Measure", "TumorVolume_cm3"])
peg_ph_o6p = df.iloc[2*series_len:].sort_values(by=["Measure", "TumorVolume_cm3"])
```

```python
plt.figure(figsize=(8, 5))

plt.plot(pbs["Measure"], pbs["TumorVolume_cm3"], label="PBS", marker='o')
plt.plot(peg_o6p["Measure"], peg_o6p["TumorVolume_cm3"], label="PEG_O6p", marker='s')
plt.plot(peg_ph_o6p["Measure"], peg_ph_o6p["TumorVolume_cm3"], label="PEG_pH_O6p", marker='^')

plt.title("Tumor Growth Over Time (Three Treatments)")
plt.xlabel("Day (Measure)")
plt.ylabel("Tumor Volume (cm³)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print("PBS:\n", pbs.describe())
print("PEG_O6p:\n", peg_o6p.describe())
print("PEG_pH_O6p:\n", peg_ph_o6p.describe())
```

```python
print("Final tumor volume (last Measure):")
print("PBS:", pbs["TumorVolume_cm3"].iloc[-1])
print("PEG_O6p:", peg_o6p["TumorVolume_cm3"].iloc[-1])
print("PEG_pH_O6p:", peg_ph_o6p["TumorVolume_cm3"].iloc[-1])
```

```python
plt.figure(figsize=(8, 5))

plt.plot(pbs['Measure'], pbs['TumorVolume_cm3'], label='PBS', color='gray', marker='o')
plt.plot(peg_o6p['Measure'], peg_o6p['TumorVolume_cm3'], label='PEG_O6p', color='teal', marker='s')
plt.plot(peg_ph_o6p['Measure'], peg_ph_o6p['TumorVolume_cm3'], label='PEG_pH_O6p', color='red', marker='^')

plt.title("Tumor Growth Over Time")
plt.xlabel("Days")
plt.ylabel("Tumor Volume (cm³)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

📊 What this graph shows:
This plot compares tumor growth over time in mice that were given different treatments:

PBS – just a control, no drug (gray line).

PEG_O6p – a DNA robot without the pH switch (teal line).

PEG_pH_O6p – the smart DNA robot that activates only in acidic tumors (red line).

The x-axis shows time (in days),
the y-axis shows tumor size (in cm³).
